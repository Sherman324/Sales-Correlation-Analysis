```python
# importing libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = 12,8
```

```python
#importing the dataset

df = pd.read_csv(r'C:\Users\Sherman\Desktop\Data Science\Programs\DATASETS\Superstore.csv')
df.head()
```

### BASIC DATA CLEANING 

```python
# Checking if there is need fro data cleaning (missing values)
# Using for loop to run through all the columns at once

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# checking data types

df.dtypes
```

```python
# arranging data by highest sales

df.sort_values(by=['Sales'], inplace=False, ascending=False)
```

```python
# Displaying the whole dataset (makes it essy to browse though)
#pd.set_option('display.max_rows',None)
```

```python
# checking for duplicates
df['Order ID'].drop_duplicates().sort_values(ascending=False)
```

### Correlation analysis

```python
# (Thought) Unit price, quantity ordered and sales has high correlation with profit
```

```python
# Scatter plot of profit vs sales
# basic visualization

plt.scatter(x=df['Sales'], y=df['Profit'])

plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

plt.show()
```

```python
# Regression plot

sns.regplot(x='Sales', y='Profit', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Looking at the correlation

df.corr(method='pearson') # pearson , kendall , spearman
```

```python
# Visualizing the correlation matrix

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot =True)

plt.title('Correlation matrix for numeric features')
plt.xlabel('Sales features')
plt.ylabel('Sales features')

plt.show()
```

```python
# Analysis on correlation of city vs profits
# Changing all columns with objects to strings using for loop

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized        
```

```python
# correlation analysis for changes columns

df_numerized.corr(method='pearson')
```

```python
# rearranging the corr in a more easier way

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.1]
high_corr
```
